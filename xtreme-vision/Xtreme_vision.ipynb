## Data preparation - YoloV4

```
import shutil
import json
import xtreme_vision
import os
```

```
label = {"Signature": 0}

json_list = os.listdir('data/json')
images_list = os.listdir('data/images')
```

```
final_list = []
for x in images_list:
    if x.split('.')[0] + '.json' in json_list:
        final_list.append(x)
```

```
!rm -r data/train/
!mkdir data/train/
```

```
for file in final_list:
    fil = file.split('.')[0] + '.json'
    json_file = os.path.join('data/json/', fil)
    json_load = open(json_file)
    json_l = json.load(json_load)
    w = json_l['size']['width']
    h = json_l['size']['height']
    f = open(os.path.join('data/train/', file.split('.')[0] + '.txt'), 'w')
    if type(json_l['object']) == list:
        for x in json_l['object']:
            if w !=0 and h !=0:
                xmin = x['bndbox']['xmin']/w
                xmax = x['bndbox']['xmax']/w
                ymin = x['bndbox']['ymin']/h
                ymax = x['bndbox']['ymax']/h
                lab = label[x['name']]

                line = str(lab) + ' ' + str(xmin) + ' ' + str(xmax) + ' ' + str(ymin) + ' ' + str(ymax)
                f.write(line+'\n')
                shutil.copy(os.path.join('data/images/', file.split('.')[0] + '.png'), os.path.join('data/train/', file.split('.')[0] + '.png'))
    else:
        if w !=0 and h !=0:
            xmin = json_l['object']['bndbox']['xmin']/w
            xmax = json_l['object']['bndbox']['xmax']/w
            ymin = json_l['object']['bndbox']['ymin']/h
            ymax = json_l['object']['bndbox']['ymax']/h
            lab = label[json_l['object']['name']]

            line = str(lab) + ' ' + str(xmin) + ' ' + str(xmax) + ' ' + str(ymin) + ' ' + str(ymax)
            f.write(line+'\n')
            shutil.copy(os.path.join('data/images/', file.split('.')[0] + '.png'), os.path.join('data/train/', file.split('.')[0] + '.png'))
    f.close()
```

```
for file in final_list:
  fil = file.split('.')[0] + '.json'
  json_file = os.path.join('data/json/', fil)
  json_load = open(json_file)
  json_l = json.load(json_load)
  w = json_l['size']['width']
  h = json_l['size']['height']
  if type(json_l['object']) != list:
    f = open(os.path.join('data/train/', file.split('.')[0] + '.txt'), 'w')
    if w !=0 and h !=0:
      xmin = round(json_l['object']['bndbox']['xmin']/w, 4)
      xmax = round(json_l['object']['bndbox']['xmax']/w, 4)
      ymin = round(json_l['object']['bndbox']['ymin']/h, 4)
      ymax = round(json_l['object']['bndbox']['ymax']/h, 4)
      lab = label[json_l['object']['name']]

      line = str(lab) + ' ' + str(xmin) + ' ' + str(xmax) + ' ' + str(ymin) + ' ' + str(ymax)
      f.write(line)
      shutil.copy(os.path.join('data/images/', file.split('.')[0] + '.png'), os.path.join('data/train/', file.split('.')[0] + '.png'))
  
  f.close()
```

```
file_names = os.listdir('data/train/')
png_files = []
for x in file_names:
    if x.split('.')[1] == 'png':
        png_files.append( x)
png_files = sorted(png_files)     
names = open('names.txt','w')
for x in png_files:
    names.write(x + "\n")
names.close()
```

## Model training - Yolov4

```
from xtreme_vision.Detection.Custom import Train_TinyYOLOv4
```

```
model = Train_TinyYOLOv4()
model.create_model(classes_path='classes.names',
                   input_size=640,
                   batch_size=2)

model.load_data(train_images_file = 'names.txt',
                train_img_dir='data/train/')

model.train(epochs=15,
            lr=1e-4,
            steps_per_epoch=400)
```

## Model Inference - TinyYoloV4

```
from xtreme_vision.Detection import Object_Detection
from PIL import Image

clf = Object_Detection()
clf.Use_TinyYOLOv4(weights_path='model_weights/yolov4-tiny-15.weights',
                   classes_path='classes.names',
                   input_shape=640,
                   iou = 0.1)
clf.Detect_From_Image(input_path='vxukS.jpg', 
                      output_path='out.jpg')
Image.open('out.jpg')
```
